**Задание 1.** Реализовать алгоритм блочного шифрования в режиме обратной связи по шифротексту на основе скиталы 

```python
import numpy as np
import time
```

```python
import secrets
import string
# Функция генерации символьной строки заданного размера
def generate_random_string_rus(length):
    rus_letters_and_digits = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ" + " " + "1234567890" # исходная строка из букв русского алфавита, цифр и пробела
    rand_string = ''.join(secrets.choice(
        rus_letters_and_digits) for i in range(length)) # функция secrets.choice() возвращает случайно выбранный элемент из последовательности
                                                    # таким образом получаем строку заданной длины
    return rand_string
```

```python
import secrets
import string

def generate_random_string_eng(length):
    eng_letters = "ABCDEFG" 
    rand_string = ''.join(secrets.choice(
    eng_letters) for i in range(length)) # функция secrets.choice() возвращает случайно выбранный элемент из последовательности
                                                    # таким образом получаем строку заданной длины
    return rand_string
```

```python
# Функция шифрования
def block_encryption(source_txt, rnd_str, n, m):
  blocks_txt = []
  start = time.time() # время начала отсчета
  for x in range (0, len(source_txt), n*m): # цикл для генерации списка с нарезками текста на блоки равной длинны 
    if (x+n*m) > len(source_txt): 
      source_txt = source_txt.ljust(x+n*m, ' ') # добавляем недостающие пробелы в конце строки 
    blocks_txt.append(source_txt[x:x+n*m])
  print("Блоки исходного текста:", blocks_txt)
  lst = np.array(list(rnd_str))
  mass = lst.reshape(n, m)
  mass_tr = mass.transpose() # транспонирование матрицы
  print(mass_tr)
  Skitala_block = ""
  for row in mass_tr:
    for ch in row:
      Skitala_block += ch
  block_encrypt = ""
  for block in blocks_txt:
    for i, ch in enumerate(block):
      block_encrypt += chr(ord(ch) ^ ord(Skitala_block[i])) # первый зашифрованный блок
    txt_encrypt = ""
    txt_encrypt += block_encrypt
    print(block_encrypt)
    break

  for i in range(1, len(blocks_txt)):
    lst = np.array(list(block_encrypt))
    mass = lst.reshape(n, m)
    mass_tr = mass.transpose() # транспонирование матрицы
    Skitala_block = ""
    for row in mass_tr:
      for ch in row:
        Skitala_block += ch
    print(mass_tr)
    block_encrypt = ""
    for j in range(len(blocks_txt[i])):
      block_encrypt += chr(ord(blocks_txt[i][j]) ^ ord(Skitala_block[j]))
    print(block_encrypt)
    txt_encrypt += block_encrypt
  end = time.time() # время конца отсчета
  print("Время выполнения шифрования:", end-start)
  return txt_encrypt
```

```python
input_text = generate_random_string_eng(30)
print('Исходный текст:', input_text)
n = 4 # ключ Скиталла (диаметр палочки, кол-во строк в матрице)
m = 3  # кол-во столбцов
rnd_str = generate_random_string_eng(n*m)
print('Вектор инициализации:', rnd_str)
txt_encrypt = block_encryption(input_text, rnd_str, n, m)
print("Зашифрованный текст:", txt_encrypt)
```

**Задание 2.** Реализовать обратную операцию расшифровки: см. схему на рисунке 1 (из-за обратимости операции XOR для этого при анализе схемы достаточно отразить направление вертикальных стрелок между c1 и m1, c2 и m2 и так далее).

```python
# Функция расшифровки
def block_decryption(txt_encrypt, rnd_str, n, m):
  start = time.time() # время начала отсчета 
  block_encrypt = [txt_encrypt[x:x+n*m] for x in range (0, len(txt_encrypt), n*m)] #генератор списка, использующий срезы для нарезки зашифрованного текста на блоки равной длинны
  for i,block in enumerate(block_encrypt):
    lst = np.array(list(rnd_str))
    print(lst)
    mass = lst.reshape(n, m)
    mass_tr = mass.transpose() # транспонирование матрицы
    print(mass_tr)
    Skitala_block = ""
    for row in mass_tr:
      Skitala_block += ''.join(map(str, row)) # к вектору инициализации применяется алгоритм Скиталы
    print(Skitala_block)
    block_decrypt = ""
    txt_decrypt = ""
    for j, ch in enumerate(Skitala_block):
      block_decrypt += chr(ord(ch) ^ ord(block_encrypt[i][j])) # получили первый блок исходного сообщения
    txt_decrypt += block_decrypt
    print("Расшифрованный блок:", block_decrypt)
    break

  for i in range(0, len(block_encrypt)-1): # обход блоков зашифрованного текста, начиная с индекса 0
    lst = np.array(list(block_encrypt[i]))
    print(lst)
    mass = lst.reshape(n, m)
    mass_tr = mass.transpose() # транспонирование матрицы
    print(mass_tr)
    Skitala_block = ""
    for row in mass_tr:
      Skitala_block += ''.join(map(str, row)) # из транспонированной к первоначальному виду матрицы получаю строку, которая изначально была получена операцией XOR
    print(Skitala_block)
    block_decrypt = ""
    for j, ch in enumerate(Skitala_block):
      block_decrypt += chr(ord(ch) ^ ord(block_encrypt[i+1][j])) # получили следующий блок исходного сообщения
    txt_decrypt += block_decrypt
    print("Расшифрованный блок:", block_decrypt)
  end = time.time() # время конца отсчета
  print("Время выполнения расшифровки:", end-start)
  return txt_decrypt
```

```python
txt_decrypt = block_decryption(txt_encrypt, rnd_str, n, m)
print("Расшифрованный текст:", txt_decrypt)
```

```python
def ReadingFile (path): # функция чтения файла
  #txt = []
  txt = ''
  f = None
  try:
      f = open(path, 'r', encoding="utf-8") #Открыть файл для чтения
      for ln in f: #Обход строк файла и добавление в список
          #txt.extend(ln) # Добавить к списку
          txt += ln # символ окончания строки заменяем на пробел
  except Exception as e:
      print("Ошибка при работе с файлом:", e)
  finally:
      if f:  # Если файл не открылся, значит 'f' == None и закрывать его не нужно
          f.close()
  return txt
```

```python
from math import log2
from collections import Counter
# Функция подсчета энтропии
def entropy (txt):
    dic = Counter(txt) # словарь с числом появлений каждого символа текста
    total = sum(dic.values()) # длину строки можно рассчитать как сумму всех частот символов
    
    return round(sum(val / total * log2(total / val) for val in dic.values()), 5)
```

**Контрольные вопросы:**
**Чем режимы шифрования из лаб. 6 и лаб. 7 отличаются друг от друга?** В режиме CBC необходимо выполнить операцию шифрования после операции XOR каждого блока открытого текста с предыдущим блоком зашифрованного текста. В отличие от этого, режим CFB напрямую не шифрует открытый текст. Вместо этого предыдущий зашифрованный блок повторно шифруется с помощью секретного ключа и XOR с открытым блоком для получения зашифрованного блока.
В режиме CBC в отличии от режима CFB для расшифрования требуется провести компазиционно обратное к Ek(алгоритму шифрования) преобразование.
**В чем схожи?** 
В таких режимах шифрования, как CBC и CFB, на вход, при шифровании первого блока, подается вектор инициализации. Причём как отправитель, так и получатель должны иметь один и тот же вектор инициализации. В обоих этих режимах это значение должно быть непредсказуемым.
Также в обоих шифрах используется способ, при котором одинаковые незашифрованные блоки преобразуются в различные зашифрованные. Для этого в качестве входа алгоритма используется результат применения операции XOR к текущему незашифрованному блоку и предыдущему зашифрованному блоку.
**Каковы плюсы и минусы каждого?** 
*Минусы:*
1. В режиме CBC незашифрованный текст обрабатывается последовательно, блок за блоком. Блоки дролжны быть определенной длинны. Если сообщение длиннее, чем длина блока соответствующего алгоритма, то оно разбивается на блоки соответствующей длины, причем последний блок дополняется в случае необходимости фиксированными значениями. 
2. В режиме CBC и CFB в случае ошибки передачи она (эта ошибка) передается в несколько последующих блоков из-за эффекта сцепления.
3. В обоих режимах чтобы расшифровать каждый следующий блок нужен зашифрованный открытый текст. Без доступного блока открытого текста не получится сгенерировать следующий блок шифрования.
*Плюсы:*
1. Режим CFB имеет преумущества поточного алгоритма шифрования. Это устраняет необходимость разбивать сообщение на целое число блоков достаточно большой длины, следовательно, он может работать в реальном времени. Таким образом, если передается поток символов, каждый символ (несколько битов) может шифроваться и передаваться сразу, с использованием символьно ориентированного режима блочного алгоритма шифрования.
2. В обоих режимах есть преимущества блочного шифрования - зашифрованный текст будет той же длинны, что и исходный.

**В каких ситуациях лучше применять один или второй?**
Алгоритм CFB более предпочтителен для реализации на более простых устройствах, т.к. операцию Ek (алгоритм щифрования) нет необходимости реализовывать для расшифрования, он работает одинаково как для шифрования и расшифрования. Этот режим более прост в реализации. CFB используется при потокоориентированной передаче. CBC используется при блоориентированной передаче
